```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3-Q1

lower_bound = Q1-1.5*(IQR)
upper_bound = Q3+1.5*(IQR)

lower_bound , upper_bound
```

```python
data = data[~((data>upper_bound)|(data<lower_bound))]
```

```python
data.shape
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
data_encoded = data[["dept","salary"]].apply(le.fit_transform)
```

```python
data_encoded
```

```python
data = data.drop(['dept','salary'] , axis = 1)
```

```python
data
```

```python
data_ecoded = pd.concat([data , data_encoded] , axis = 1)
```

```python
data_ecoded
```

```python
data_ecoded.isnull().sum()
```

```python
data_ecoded = data_ecoded.fillna(data_ecoded[['promotion_last_5years','Work_accident','time_spend_company']].mean())
```

```python
x = data_ecoded.drop(['satisfaction_level'] , axis = 1)
y = data_ecoded['satisfaction_level']
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.30 , random_state = 42)
```

```python
model_linear = LinearRegression()
```

```python
model_linear.fit(x_train , y_train)
```

```python
y_pred = model_linear.predict(x_test)
```

```python
y_pred
```

```python
pd.DataFrame({'actual':y_test ,'predicted':y_pred })
```

```python
mse = np.mean((y_test-y_pred)**2)
```

```python
mse
```

```python
rmse = np.sqrt(mse)
```

```python
rmse
```

```python
y_ridge = model_ridge.predict(x_test)
```
